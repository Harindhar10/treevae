```python
import torch
from torch.utils.data import Dataset

class CustomDataset(Dataset):
    def __init__(self, data, num_classes):
        """
        Args:
            data (torch.Tensor): A 3D tensor of shape (N, H, W), where
                                 N is the number of samples, H and W are the dimensions of each 2D tensor.
            num_classes (int): Number of classes for generating random labels.
        """
        self.data = data
        self.num_classes = num_classes
        self.targets = torch.randint(0, num_classes, (data.size(0),))  # Random labels for each sample

    def __len__(self):
        return self.data.size(0)  # Number of samples

    def __getitem__(self, idx):
        # Extract 2D tensor and reshape to triple square brackets format
        tensor_2d = self.data[idx]  # Add extra brackets: [[[tensor]]]
        tensor_2d_reshaped = torch.reshape(tensor_2d,(1,784))
        # Fetch the corresponding random label
        label = self.targets[idx]
        
        return tensor_2d_reshaped, label

# Example usage
N, H, W = 10, 28, 28  # 10 samples of 28x28 tensors
data = torch.rand(N, H, W)  # Random 3D tensor
num_classes = 5  # Assume 5 random classes

# Create the dataset
dataset = CustomDataset(data, num_classes)
print('dataset targets',dataset.targets)
# Access samples
for i in range(3):
    tensor, label = dataset[i]
    print(f"Sample {i}:")
    print(f"Tensor shape: {tensor.shape}")  # Shape should be [[[H, W]]]
    print(f"Label: {label}")
```
